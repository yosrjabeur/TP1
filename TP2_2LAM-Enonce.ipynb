### Simulation
Ecrire une foction `sim` qui simule le jeux et qui prend comme variable le nombre de pas $n$ nécessazire pour atteindre la cible

```python
def sim(S):
    NA=0
    S=10
    c=0
## sortir du prison ##
    temp=Roll()
    NA+=1
    while(temp!=6):
        temp=Roll()
        NA+=1
    A=1
    print("sortir aprés {} coups ".format(NA))
## parcours vers la cible
    while(A!=S):
        temp=Roll()
        if (temp+A<S):
            A=A+temp
      #  NA+=1
        elif(temp+A>S):
            continue
        else:
            A=S
    return NA 
#le nombre de cout necessaire pour atteindre la position cible
        
    
```

```python
#simulation 10000 fois n=30
L=[sim(30) for i in range(10000)]
Moy=sim(L)/10000
print('stimation de la moyenne de X_30={}'.format(Moy))
```
